```python
import pandas as pd
import numpy as np
import csv

# Definio do DataFrame em uma varivel

df = pd.read_csv('dados4.csv')

# Obteno da moda da coluna "age"

dfMode = df['age'].mode().iat[0]

# Este bloco try except tenta primeiramente substituir os valores menores que 1 pela moda, e em seguida substitui os valores vazios.

try:

    df.loc[df['age'] < 1, 'age'] = dfMode

    df['age'] = df['age'].fillna(dfMode)
    
# Caso haja algum erro, o seguinte erro ser printado no console:
    
except:
    
    print('No foi possvel alterar os valores invlidos ou nulos do DataFrame.')
    
    
# Este bloco try except tenta abrir o arquivo "Resposta01.txt" e em seguida salva os valores do "dados4.csv" (caso o arquivo ainda no exista, ele ser criado).
    
try:

    with open('Resposta01.txt', 'w') as firstExercise:
        firstExercise.write(df.to_string())
        
# Caso haja algum erro, o seguinte erro ser printado no console:
        
except:
    
    print('No foi possvel salvar os valores do DataFrame no arquivo .txt.')
    
    
# A soma  calculada e em seguida printada no console. No  necessrio um bloco try except pois no h muitas variveis que podem ocasionar um erro.

sum = df['age'].sum()
print(f'soma de todas as idades: {sum}')


# Este bloco try except conta quantas entradas h no dados4.csv. aps isso, conta quantas vezes os valores 0 e 1 aparecem, para depois calcular a porcentagem de cada um e em seguida armazen-los em uma varivel. Aps isso, o grfico de pizza  gerado.

try:

    totalEntries = len(df)

    survivedCount = df['survived'].value_counts()

    dfSurvivedPercentage = (survivedCount / totalEntries) * 100

    dfSurvivedPercentage.plot.pie(labels=['Sobreviveu', 'No sobreviveu'], autopct='%1.1f%%')
    
# Caso haja algum erro, o seguinte erro ser printado no console:
    
except:
    
    print('No foi possvel gerar o grfico de sobreviventes.')
    
# O grfico de disperso  gerado de acordo com as colunas 'fare' e 'age'.
    
df.plot.scatter(x='fare', y='age')
```
