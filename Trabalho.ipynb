```python
import pandas as pd
import numpy as np
import csv

# Definição do DataFrame em uma variável

df = pd.read_csv('dados4.csv')

# Obtenção da moda da coluna "age"

dfMode = df['age'].mode().iat[0]

# Este bloco try except tenta primeiramente substituir os valores menores que 1 pela moda, e em seguida substitui os valores vazios.

try:

    df.loc[df['age'] < 1, 'age'] = dfMode

    df['age'] = df['age'].fillna(dfMode)
    
# Caso haja algum erro, o seguinte erro será printado no console:
    
except:
    
    print('Não foi possível alterar os valores inválidos ou nulos do DataFrame.')
    
    
# Este bloco try except tenta abrir o arquivo "Resposta01.txt" e em seguida salva os valores do "dados4.csv" (caso o arquivo ainda não exista, ele será criado).
    
try:

    with open('Resposta01.txt', 'w') as firstExercise:
        firstExercise.write(df.to_string())
        
# Caso haja algum erro, o seguinte erro será printado no console:
        
except:
    
    print('Não foi possível salvar os valores do DataFrame no arquivo .txt.')
    
    
# A soma é calculada e em seguida printada no console. Não é necessário um bloco try except pois não há muitas variáveis que podem ocasionar um erro.

sum = df['age'].sum()
print(f'soma de todas as idades: {sum}')


# Este bloco try except conta quantas entradas há no dados4.csv. após isso, conta quantas vezes os valores 0 e 1 aparecem, para depois calcular a porcentagem de cada um e em seguida armazená-los em uma variável. Após isso, o gráfico de pizza é gerado.

try:

    totalEntries = len(df)

    survivedCount = df['survived'].value_counts()

    dfSurvivedPercentage = (survivedCount / totalEntries) * 100

    dfSurvivedPercentage.plot.pie(labels=['Sobreviveu', 'Não sobreviveu'], autopct='%1.1f%%')
    
# Caso haja algum erro, o seguinte erro será printado no console:
    
except:
    
    print('Não foi possível gerar o gráfico de sobreviventes.')
    
# O gráfico de dispersão é gerado de acordo com as colunas 'fare' e 'age'.
    
df.plot.scatter(x='fare', y='age')
```
